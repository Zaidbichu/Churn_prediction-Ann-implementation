```python
import tensorflow as tf
from tensorflow.keras import models
import pickle
import keras
import pandas as pd
import numpy as np
```

```python
with open('model.keras','rb') as file:
    model = keras.models.load_model('model.keras')

#load the label encoder pickle file
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

#load tje one hot encoder pickle file
with open('one_hot_encoder_.pkl','rb') as file:
    one_hot_enco=pickle.load(file)

#load the standadization pickle file
with open('scalar.pkl','rb') as file:
    scalar=pickle.load(file)
```

```python
input_data={
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
geo_encode=one_hot_enco.transform([[input_data['Geography']]]).toarray()
geo_encoder_df=pd.DataFrame(geo_encode,columns=one_hot_enco.get_feature_names_out(['Geography']))
geo_encoder_df
```

```python
input_data=pd.DataFrame([input_data])
```

```python
input_data['Gender']=label_encoder_gender.transform(input_data['Gender'])
```

```python
input_data
```

```python
input_data=pd.concat([input_data.drop('Geography',axis=1),geo_encoder_df],axis=1)
```

```python
input_data
```

```python
input_scaled=scalar.transform(input_data)
```

```python
prediction=model.predict(input_scaled)
```

```python
prediction_proba=prediction[0][0]
```

```python
prediction_proba
```

```python
if prediction_proba>0.5:
    print('the customer is likely to churn')
else:
    print('the customer is not likely to churn')
```
