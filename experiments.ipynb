```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
##removing the irevelant columns
data.drop(['RowNumber',	'CustomerId',	'Surname'],axis=1,inplace=True)
```

```python
##performing label encodinf on the gender column
lab_encoder=LabelEncoder()
data['Gender']=lab_encoder.fit_transform(data['Gender'])
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder=OneHotEncoder()
geo_encoder=one_hot_encoder.fit_transform(data[['Geography']]).toarray()
```

```python
one_hot_encoder.get_feature_names_out(['Geography'])
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder,columns=one_hot_encoder.get_feature_names_out(['Geography']))
geo_encoder_df
```

```python
## combine the one hot encoded data to the original data
data=pd.concat([data.drop('Geography',axis=1),geo_encoder_df],axis=1)
```

```python
data.head()
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(lab_encoder,file)
with open('one_hot_encoder_.pkl','wb') as file:
    pickle.dump(one_hot_encoder,file)
```

```python
data
```

```python
##identifying the dependent and independent features
x=data.drop('Exited',axis=1)
y=data['Exited']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.32, random_state=42)
```

```python
scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

```python
##ann impleentation
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
(x_train.shape[1],)
```

```python
##build our ann model
model=Sequential(
    [
        Dense(32,activation='relu',input_shape=(x_train.shape[1],))
    ]
)
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
loss
```

```python
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir="logs/fit/"+datetime.datetime.now().strftime("%Y%M%D-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
import tensorflow
model=Sequential(
    [
        Dense(32,activation='relu',input_shape=(x_train.shape[1],)),
        Dense(1,activation='sigmoid')
    ]
)
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=opt,loss='BinaryCrossentropy',metrics=['accuracy'])
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.keras')
```
